## Red neuronal

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_neural_net(layers):
    """
    layers: lista que contiene el n√∫mero de neuronas en cada capa
    """
    G = nx.DiGraph()  # Crear un grafo
    n_layers = len(layers)
    vertical_spacing = 1  # Espacio vertical entre neuronas
    horizontal_spacing = 2  # Espacio horizontal entre capas
    node_counter = 0  # Contador para los identificadores de los nodos

    for i in range(n_layers):
        layer_size = layers[i]
        layer_positions = []
        x = i * horizontal_spacing  # Coordenada x para la capa actual
        # Ajuste para centrar las neuronas verticalmente
        y_offset = (max(layers) - layer_size) / 2.0
        for j in range(layer_size):
            y = (j + y_offset) * vertical_spacing  # Coordenada 'y' para la neurona actual
            G.add_node(node_counter, pos=(x, y))  # Agregar nodo al grafo
            layer_positions.append(node_counter)
            node_counter += 1
        if i > 0:
            # Conectar neuronas de la capa anterior con la actual
            for node_origin in previous_layer_positions:
                for node_destination in layer_positions:
                    G.add_edge(node_origin, node_destination)
        previous_layer_positions = layer_positions  # Actualizar capa anterior

    pos = nx.get_node_attributes(G, 'pos')  # Obtener posiciones de los nodos
    nx.draw(G, pos, with_labels=False, node_size=500, node_color='orange', arrowsize=10)
    plt.title('Red Neuronal Full Connected')
    plt.axis('off')  # Ocultar ejes
    plt.show()

# Configuraci√≥n de la red neuronal con 2 capas ocultas de 5 neuronas cada una
layers = [3, 5, 5, 5, 2]  # 3 neuronas en la capa de entrada, 5 en cada capa oculta, 1 en la capa de salida
draw_neural_net(layers)
```
